```python
import math
```

```python
k=17 # DD homozygous dominant
m=18 #Dd heterozygous dominant
n=24 #dd homozygous recessive
total=k+m+n #total number of plants
#below is the probability of occurence of the dominant phenotype for each cross
kxk_cross=1 #DDXDD
kxm_cross=1 #DDXDd
kxn_cross=1 #DDXdd
mxm_cross=0.75 #DdXDd
mxk_cross=1 #DdXDD
mxn_cross=0.5 #DdXdd
nxn_cross=0 #ddXdd
nxm_cross=0.5 #ddXDd
nxk_cross=1 # ddXDD

prob_DDxDd=((k*m)/math.comb(total,2))*kxm_cross
prob_DDxDD=(math.comb(k,2)/math.comb(total,2))*kxk_cross
prob_DdxDd=(math.comb(m,2)/math.comb(total,2))*mxm_cross
prob_DDxdd=((k*n)/math.comb(total,2))*kxn_cross
prob_Ddxdd=((m*n)/math.comb(total,2))*mxn_cross
prob_ddxdd=(math.comb(n,2)/math.comb(total,2))*nxn_cross

dom_freq=prob_DDxDd+prob_DDxDD+prob_DdxDd+prob_DDxdd+prob_Ddxdd

print(dom_freq)
```
