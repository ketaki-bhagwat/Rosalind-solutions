```python
with open(r"D:\Python projects\rosalind_prot.txt","r") as files:
    seq=files.read()
    RNA=seq.strip()
codons=[]
for i in range (0,len(RNA),3):
    s=RNA[i:i+3]
    codons.append(s)
    codon_lib=[]
with open (r"D:\Python projects\Codon table.txt","r") as file:
    for line in file:
        n=line.strip()
        c=n.split("\t")
        codon_lib.append(c)
del codon_lib[0]
codon_table=dict(codon_lib)
protein=[]
for i in codons:
    if i in codon_table:
        amino_acid=codon_table[i]
        if amino_acid =="stop":
            break
        protein.append(amino_acid)

translation="".join(protein)
print(translation)    
```
