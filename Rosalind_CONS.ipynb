```python
with open(r"D:\Python projects\rosalind_cons.txt", "r") as file:
    data=file.read()
    all_sequences=[]
    current_sequence=""
    for line in data.splitlines():
        if line.startswith(">"):
            if len(current_sequence)>0:
                all_sequences.append(current_sequence)
                current_sequence=""     
        else:
            current_sequence+=line
    all_sequences.append(current_sequence)
for sequence in all_sequences:
    length=len(sequence)
    if length!=len(all_sequences[0]):
        print("Sequences are not of the same length.")
        exit()
    else:
        matrix=[list(sequence) for sequence in all_sequences]
#print (matrix)

nucleotides= "ATGC"
profile={}
for letter in nucleotides:
    profile[letter]=[0]*len(all_sequences[0])

for row in matrix:
    for position,letter in enumerate(row):
        profile[letter][position]+=1
#print(profile)

consensus=[]
for position in range(len(profile['A'])):
    max_count=0
    base_letter=None
    for letter in "ATGC":
        current_count=profile[letter][position]
        if current_count>max_count:
            max_count=current_count
            base_letter=letter
    consensus.append(base_letter)
print("".join(consensus))
for key,value in profile.items():
    print(key+":", " ".join(map(str,value)))
```
